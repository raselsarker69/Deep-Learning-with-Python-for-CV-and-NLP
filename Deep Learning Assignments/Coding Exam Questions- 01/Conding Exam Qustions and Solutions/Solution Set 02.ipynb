### Question 1(b): Control Flow and Functions

```python

# Solution for Question 1(b)
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Print all prime numbers between 2 and 100
for num in range(2, 101):
    if is_prime(num):
        print(num, end=" ")
```

### Question 2(b): Dictionary and Set Operations

```python

# Solution for Question 2(b)
english_students = {"Emma", "Olivia", "Ava", "Isabella", "Sophia"}
math_students = {"Olivia", "Ava", "Mia", "Charlotte", "Amelia"}

# Students in both courses
both_courses = english_students.intersection(math_students)
print("Students in both courses:", both_courses)

# Students in English but not in Math
only_english = english_students.difference(math_students)
print("Students in English but not in Math:", only_english)

# Create a dictionary of students and their courses
all_students = english_students.union(math_students)
student_courses = {student: [] for student in all_students}
for student in english_students:
    student_courses[student].append("English")
for student in math_students:
    student_courses[student].append("Math")
print("Student courses:", student_courses)
```

### Question 4(a): Data Loading and Cleaning

```python

# Solution for Question 4(a)
import pandas as pd

df = pd.read_csv('sales.csv')
print("First five rows:\n", df.head())

# Handling missing values
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['UnitPrice'] = df['UnitPrice'].fillna(df['UnitPrice'].median())

# Convert 'OrderDate' to datetime and extract year
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Year'] = df['OrderDate'].dt.year
print("Updated DataFrame:\n", df.head())
```

### Question 4(b): Data Manipulation and Analysis

```python

# Solution for Question 4(b)
# Calculate total sales
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
print("DataFrame with TotalSales column:\n", df.head())

# Top 3 products by total sales
top_products = df.groupby('Product')['TotalSales'].sum().nlargest(3)
print("Top 3 products by total sales:\n", top_products)

# Average unit price by category
avg_price_by_category = df.groupby('Category')['UnitPrice'].mean()
print("Average unit price by category:\n", avg_price_by_category)
```

### Question 4(c): Advanced Filtering

```python

# Solution for Question 4(c)
filtered_orders = df[(df['Year'] == 2022) & 
                     (df['Category'] == 'Electronics') & 
                     (df['Quantity'] > 5)]
print("Filtered orders:\n", filtered_orders)
```

### Question 5(a): Matplotlib Visualization

```python

# Solution for Question 5(a)
import matplotlib.pyplot as plt

# Pie chart for total sales by category
sales_by_category = df.groupby('Category')['TotalSales'].sum()
sales_by_category.plot(kind='pie', autopct='%1.1f%%', title='Total Sales by Category')
plt.ylabel('')  # Hide y-axis label
plt.legend(title='Categories')
plt.show()
```

### Question 5(b): Seaborn Visualization

```python

# Solution for Question 5(b)
import seaborn as sns

# Heatmap for correlations
correlation = df[['Quantity', 'UnitPrice', 'TotalSales']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

### Question 6: Hands-On Project

```python

# Solution for Question 6
# Load population data
population_data = pd.read_csv('population_data.csv')

# Calculate population density for 2020
population_data_2020 = population_data[population_data['Year'] == 2020]
population_data_2020['PopulationDensity'] = population_data_2020['Population'] / population_data_2020['Area']

# Top 5 countries by population density
top_density_countries = population_data_2020.nlargest(5, 'PopulationDensity')
print("Top 5 countries by population density:\n", top_density_countries)

# Plotting
plt.bar(top_density_countries['Country'], top_density_countries['PopulationDensity'])
plt.title('Top 5 Countries by Population Density (2020)')
plt.xlabel('Country')
plt.ylabel('Population Density')
plt.xticks(rotation=45)
plt.savefig('population_density.png')
plt.show()
```

### Question 7: Capstone Coding Challenge

```python

# Solution for Question 7
# Read and clean text
with open('article.txt', 'r') as file:
    text = file.read().lower()

# Remove punctuation
import string
text = text.translate(str.maketrans('', '', string.punctuation))

# Count word frequencies
from collections import Counter
word_counts = Counter(text.split())

# Create DataFrame
word_freq_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Frequency']).sort_values(by='Frequency', ascending=False)

# Top 15 most frequent words
top_words = word_freq_df.head(15)

# Plotting
sns.barplot(x='Frequency', y='Word', data=top_words, palette='viridis')
plt.title('Top 15 Most Frequent Words')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.show()
```
