### Question 1(a): Conditional Statements and Loops

```python

# Solution for Question 1(a)
input_list = list(map(int, input("Enter a list of integers separated by spaces: ").split()))
even_numbers = [num for num in input_list if num % 2 == 0]
odd_sum = sum(num for num in input_list if num % 2 != 0)

print(f"Even numbers: {even_numbers}")
print(f"Sum of odd numbers: {odd_sum}")
```

### Question 1(b): Functions and Scope

```python

# Solution for Question 1(b)
def calculate_grade(scores):
    grades = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}
    for score in scores:
        if score >= 90:
            grades["A"] += 1
        elif score >= 80:
            grades["B"] += 1
        elif score >= 70:
            grades["C"] += 1
        elif score >= 60:
            grades["D"] += 1
        else:
            grades["F"] += 1
    return grades

# Example usage
student_scores = [95, 85, 72, 67, 45, 88]
result = calculate_grade(student_scores)
print("Grade distribution:", result)
```

### Question 2(a): List and Tuple Manipulations

```python

# Solution for Question 2(a)
products = [("Laptop", 1200), ("Smartphone", 800), ("Tablet", 400), ("Monitor", 300)]

# Sort without using sort()
sorted_products = sorted(products, key=lambda x: -x[1])
print("Sorted products:", sorted_products)

# Convert to dictionary
products_dict = {product: price for product, price in products}
print("Products dictionary:", products_dict)
```

### Question 2(b): Set and Dictionary Operations

```python

# Solution for Question 2(b)
course_python = {"Alice": 85, "Bob": 78, "Charlie": 92, "Diana": 88}
course_data_science = {"Eve": 91, "Frank": 76, "Alice": 89, "Charlie": 95}

# Students enrolled in both courses
common_students = set(course_python.keys()).intersection(set(course_data_science.keys()))
print("Students in both courses:", common_students)

# Average scores
average_scores = {student: (course_python[student] + course_data_science[student]) / 2 for student in common_students}
print("Average scores:", average_scores)
```

### Question 3(a): Array Creation and Manipulation

```python

# Solution for Question 3(a)
import numpy as np

# Create random array
A = np.random.randint(10, 50, (3, 5))
print("Array A:", A)

# Statistics
mean_val = np.mean(A)
median_val = np.median(A)
std_dev = np.std(A)
print(f"Mean: {mean_val}, Median: {median_val}, Std Dev: {std_dev}")

# Normalize array
normalized_A = (A - np.min(A)) / (np.max(A) - np.min(A))
print("Normalized Array:", normalized_A)
```

### Question 3(b): Broadcasting and Mathematical Operations

```python

# Solution for Question 3(b)
X = np.array([1, 2, 3])
Y = np.array([[10], [20], [30]])

# Broadcasting
result = X * Y
print("Result of broadcasting:", result)

# Explanation
# Broadcasting allows the smaller array (X) to be expanded to match the dimensions of the larger array (Y).
```

### Question 4(a): Data Loading and Basic Operations

```python

# Solution for Question 4(a)
import pandas as pd

# Load dataset
df = pd.read_csv('employees.csv')

# Display first five rows and statistics
print("First five rows:", df.head())
print("Basic statistics:", df.describe())

# Filter data
filtered_df = df[(pd.to_datetime(df['JoiningDate']) > '2020-01-01') & (df['Department'] == 'Data Science')]
print("Filtered DataFrame:", filtered_df)
```

### Question 4(b): Data Aggregation and Grouping

```python

# Solution for Question 4(b)
# Average salary by department
avg_salary = df.groupby('Department')['Salary'].mean()
print("Average salary by department:", avg_salary)

# Department with highest number of employees
dept_counts = df['Department'].value_counts()
most_employees = dept_counts.idxmax()
print(f"Department with the highest number of employees: {most_employees}")
```

### Question 4(c): Handling Missing Data

```python
# Identify missing values
missing_salaries = df['Salary'].isnull().sum()
print(f"Number of missing salaries: {missing_salaries}")

# Impute missing values with the median salary of respective department
df['Salary'] = df.groupby('Department')['Salary'].apply(
    lambda x: x.fillna(x.median())
).reset_index(level=0, drop=True)

print("Missing values filled with median salary.")
```

### Question 5(a): Matplotlib Visualization

```python

# Solution for Question 5(a)
import matplotlib.pyplot as plt

# Bar chart for number of employees in each department
dept_counts = df['Department'].value_counts()
dept_counts.plot(kind='bar', color='skyblue', title='Number of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

### Question 5(b): Seaborn Visualization

```python

# Solution for Question 5(b)
import seaborn as sns

# Box plot for salaries by department
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

### Question 6: Hands-On Project

```python
import seaborn as sns
# Solution for Question 6
sales_data = pd.read_csv('sales_data.csv')

# Convert 'Date' column and set as index
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data.set_index('Date', inplace=True)

# Resample to monthly total revenue
monthly_revenue = sales_data.resample('M')['Revenue'].sum()
monthly_revenue.plot(title='Monthly Total Revenue', ylabel='Revenue', xlabel='Month')
plt.show()

# Top 3 products by revenue
top_products = sales_data.groupby('ProductID')['Revenue'].sum().nlargest(3)
sns.barplot(x=top_products.index, y=top_products.values)
plt.title('Top 3 Products by Revenue')
plt.xlabel('Product ID')
plt.ylabel('Total Revenue')
plt.show()
```

### Question 7: Capstone Coding Challenge

```python

# Solution for Question 7
weather_data = pd.read_csv('weather_data.csv')

# Moving average for temperature
weather_data['Date'] = pd.to_datetime(weather_data['Date'])
weather_data['7-Day MA'] = weather_data['Temperature'].rolling(window=7).mean()

# Plot original and moving average
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Date'], weather_data['Temperature'], label='Original Temperature', color='blue')
plt.plot(weather_data['Date'], weather_data['7-Day MA'], label='7-Day Moving Average', color='orange')
plt.legend()
plt.title('Temperature and 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```
