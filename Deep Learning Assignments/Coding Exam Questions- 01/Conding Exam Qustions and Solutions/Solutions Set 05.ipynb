### Question 1(b): Functions and Error Handling

```python

# Solution for Question 1(b)
def factorial(n):
    """
    Calculate the factorial of a non-negative integer.
    Raises ValueError if the input is negative or not an integer.
    """
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Examples
try:
    print("Factorial of 5:", factorial(5))
    print("Factorial of -3:", factorial(-3))  # This will raise ValueError
except ValueError as e:
    print(e)
```

### Question 2(b): Dictionary Operations

```python

# Solution for Question 2(b)
employees = {
    "E001": "HR",
    "E002": "IT",
    "E003": "Finance",
    "E004": "IT",
    "E005": "HR",
    "E006": "Marketing",
    "E007": "Finance",
    "E008": "IT"
}

# Map departments to employee IDs
dept_to_employees = {}
for emp_id, dept in employees.items():
    if dept not in dept_to_employees:
        dept_to_employees[dept] = []
    dept_to_employees[dept].append(emp_id)
print("Department to Employees Mapping:", dept_to_employees)

# Department with the highest number of employees
max_dept = max(dept_to_employees, key=lambda k: len(dept_to_employees[k]))
print("Department with the highest number of employees:", max_dept)
```

### Question 3(a): Array Creation and Statistical Analysis

```python

# Solution for Question 3(a)
import numpy as np

# Create a random array
data = np.random.randint(50, 150, 100)

# Reshape to (20, 5)
data_matrix = data.reshape(20, 5)

# Statistics
mean_val = np.mean(data_matrix)
median_val = np.median(data_matrix)
std_dev = np.std(data_matrix)
variance = np.var(data_matrix)

# Max value in each column
max_in_columns = np.max(data_matrix, axis=0)

print(f"Mean: {mean_val}, Median: {median_val}, Std Dev: {std_dev}, Variance: {variance}")
print("Max in each column:", max_in_columns)
```

### Question 4(a): Data Loading and Exploration

```python
# Solution for Question 4(a)
import pandas as pd

# Load dataset
customer_df = pd.read_csv('customer_data.csv')

# Display basic statistics for numeric columns
print(customer_df.describe())

# Handle missing values
# Select numeric columns only
numeric_cols = customer_df.select_dtypes(include=['number']).columns
customer_df[numeric_cols] = customer_df[numeric_cols].fillna(customer_df[numeric_cols].mean())
print("Missing values in numeric columns handled.")

# Convert Gender to numerical values
if 'Gender' in customer_df.columns:
    customer_df['Gender'] = customer_df['Gender'].map({'Male': 0, 'Female': 1})
    print("Gender column converted to numerical values.")

# Display the updated DataFrame
print(customer_df.head())
```

### Question 4(b): Data Visualization and Interpretation

```python

# Solution for Question 4(b)
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot for AnnualIncome vs SpendingScore
sns.scatterplot(x='AnnualIncome', y='SpendingScore', data=customer_df)
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()

# Histogram of Age distribution by Gender
sns.histplot(data=customer_df, x='Age', hue='Gender', kde=True, bins=10)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.show()
```

### Question 5(a): Cluster Analysis Visualization

```python

# Solution for Question 5(a)
from sklearn.cluster import KMeans

# K-Means clustering on AnnualIncome and SpendingScore
kmeans = KMeans(n_clusters=5, random_state=0).fit(customer_df[['AnnualIncome', 'SpendingScore']])
customer_df['Cluster'] = kmeans.labels_

# Scatter plot for clusters
sns.scatterplot(x='AnnualIncome', y='SpendingScore', hue='Cluster', data=customer_df, palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend(title='Cluster')
plt.show()
```

### Question 4(c): Data Segmentation

```python

# Solution for Question 4(c)
# Segment customers into age groups
def age_group(age):
    if 18 <= age <= 30:
        return 'Youth'
    elif 31 <= age <= 50:
        return 'Adult'
    elif age > 50:
        return 'Senior'

customer_df['AgeGroup'] = customer_df['Age'].apply(age_group)
print("Added AgeGroup column to the DataFrame:\n", customer_df[['Age', 'AgeGroup']].head())
```

### Question 6: Hands-On Project

```python

# Solution for Question 6
# Load energy consumption data
energy_df = pd.read_csv('energy_consumption.csv')
energy_df['Timestamp'] = pd.to_datetime(energy_df['Timestamp'])
energy_df.set_index('Timestamp', inplace=True)

# Resample to daily total energy consumption
daily_consumption = energy_df['EnergyConsumption'].resample('D').sum()

# Plot daily energy consumption
plt.figure(figsize=(10, 6))
plt.plot(daily_consumption.index, daily_consumption.values, label='Daily Energy Consumption')
plt.title('Daily Energy Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.legend()
plt.show()

# Calculate 7-day rolling average
rolling_avg = daily_consumption.rolling(window=7).mean()

# Plot rolling average
plt.figure(figsize=(10, 6))
plt.plot(daily_consumption.index, daily_consumption.values, label='Daily Energy Consumption', alpha=0.5)
plt.plot(rolling_avg.index, rolling_avg.values, label='7-Day Rolling Average', color='red')
plt.title('Daily Energy Consumption with Rolling Average')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.legend()
plt.show()
```

```python
# Solution for Question 7
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Ensure the dataset has enough observations
if len(daily_consumption) >= 14:
    # Decompose the time series
    decomposed = seasonal_decompose(daily_consumption, model='additive', period=7)
    
    # Plot components
    decomposed.plot()
    plt.show()
    
    # Forecasting (Simple Example)
    last_known = daily_consumption.iloc[-1]
    forecast = [last_known + decomposed.trend.dropna().diff().mean()] * 7
    forecast_dates = pd.date_range(start=daily_consumption.index[-1] + pd.Timedelta(days=1), periods=7)
    
    # Plot forecast
    plt.figure(figsize=(10, 6))
    plt.plot(daily_consumption.index, daily_consumption.values, label='Observed')
    plt.plot(forecast_dates, forecast, label='Forecast', linestyle='--', color='orange')
    plt.title('Energy Consumption Forecast')
    plt.xlabel('Date')
    plt.ylabel('Energy Consumption')
    plt.legend()
    plt.show()
else:
    print("Not enough data for seasonal decomposition. At least 14 observations are required.")
```
