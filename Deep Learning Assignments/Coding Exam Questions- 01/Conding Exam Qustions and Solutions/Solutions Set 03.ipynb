### Question 1(b): String Manipulation and Functions

```python

# Solution for Question 1(b)
def process_string(s):
    s = s.lower()
    vowels = "aeiou"
    num_vowels = sum(1 for char in s if char in vowels)
    num_consonants = sum(1 for char in s if char.isalpha() and char not in vowels)
    print(f"Number of vowels: {num_vowels}")
    print(f"Number of consonants: {num_consonants}")
    return s[::-1]

# Example usage
input_string = input("Enter a string: ")
reversed_string = process_string(input_string)
print(f"Reversed string: {reversed_string}")
```

### Question 2(a): Conditional Statements and Loops

```python

# Solution for Question 2(a)
# Generate numbers from 1 to 50 and apply FizzBuzz logic
for num in range(1, 51):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

### Question 2(b): Data Structures Manipulation

```python

# Solution for Question 2(b)
students_scores = {
    "Emily": [88, 92, 85],
    "John": [78, 76, 80],
    "Sarah": [90, 91, 89],
    "Michael": [65, 70, 72],
    "Jessica": [95, 98, 94]
}

# Calculate average scores and filter passed students
passed_students = {}
for student, scores in students_scores.items():
    avg_score = sum(scores) / len(scores)
    print(f"{student}: Average Score = {avg_score}")
    if avg_score >= 85:
        passed_students[student] = avg_score

print("Passed Students:", passed_students)
```

### Question 4(a): Data Loading and Preprocessing

```python

# Solution for Question 4(a)
import pandas as pd
from datetime import datetime

# Load dataset
df = pd.read_csv('employees.csv')

# Convert 'JoiningDate' to datetime format
df['JoiningDate'] = pd.to_datetime(df['JoiningDate'])

# Calculate 'Experience' in years
today = datetime.today()
df['Experience'] = (today - df['JoiningDate']).dt.days // 365
print("Updated DataFrame:\n", df.head())
```

### Question 4(b): Data Manipulation and Analysis

```python

# Solution for Question 4(b)
# Filter for specific departments and salary
filtered_df = df[(df['Department'].isin(['IT', 'HR'])) & (df['Salary'] > 70000)]

# Group by department and calculate averages
grouped_data = filtered_df.groupby('Department').agg({'Experience': 'mean', 'Salary': 'mean'})
print("Grouped Data:\n", grouped_data)
```

### Question 4(c): Data Exporting

```python

# Solution for Question 4(c)
# Export grouped data to CSV
grouped_data.to_csv('department_analysis.csv', index=False)
print("Data exported to 'department_analysis.csv'.")
```

```python
# Solution for Question 6
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel sheets into DataFrames
sales_2022 = pd.read_excel('sales_data.xlsx', sheet_name='2022_Sales')
sales_2023 = pd.read_excel('sales_data.xlsx', sheet_name='2023_Sales')

# Combine into a single DataFrame
combined_sales = pd.concat([sales_2022, sales_2023])

# Add 'TotalSales' column
combined_sales['TotalSales'] = combined_sales['Quantity'] * combined_sales['UnitPrice']

# Ensure 'Date' column is in datetime format
combined_sales['Date'] = pd.to_datetime(combined_sales['Date'])

# Top 5 products by total sales
top_products = combined_sales.groupby('Product')['TotalSales'].sum().nlargest(5)
print("Top 5 products by total sales:\n", top_products)

# Line chart for monthly total sales of the top product
top_product = top_products.idxmax()

# Filter for the top product and set 'Date' as the index
top_product_sales = combined_sales[combined_sales['Product'] == top_product].set_index('Date')

# Resample by month and calculate total sales
monthly_sales = top_product_sales.resample('M')['TotalSales'].sum()

# Plot the monthly sales
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title(f'Monthly Sales for {top_product}')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend([top_product])
plt.grid(True)
plt.show()
```

### Question 7: Capstone Coding Challenge

```python

# Solution for Question 7
# Load temperature data
temperature_data = pd.read_csv('temperature_data.csv')

# Convert 'Date' to datetime and set as index
temperature_data['Date'] = pd.to_datetime(temperature_data['Date'])
temperature_data.set_index('Date', inplace=True)

# Resample to monthly averages
temperature_data['Temperature'] = temperature_data['Temperature'].interpolate()
monthly_avg_temp = temperature_data['Temperature'].resample('M').mean()

# Plot original and monthly average temperatures
plt.figure(figsize=(10, 6))
plt.plot(temperature_data.index, temperature_data['Temperature'], label='Daily Temperature', alpha=0.7)
plt.plot(monthly_avg_temp.index, monthly_avg_temp.values, label='Monthly Average Temperature', color='red')
plt.title('Temperature Trends')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```
