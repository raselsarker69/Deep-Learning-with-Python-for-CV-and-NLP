### Question 1(a): Basic Input/Output and Data Types

```python

# Solution for Question 1(a)
numbers = list(map(int, input("Enter numbers separated by commas: ").split(',')))

# Calculate and print statistics
average = sum(numbers) / len(numbers)
maximum = max(numbers)
minimum = min(numbers)

print(f"Average: {average}")
print(f"Maximum: {maximum}")
print(f"Minimum: {minimum}")
```

### Question 1(b): Functions and Error Handling

```python

# Solution for Question 1(b)
def divide_numbers(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
        return None

# Examples
result1 = divide_numbers(10, 2)
print(f"Result of 10 / 2: {result1}")

result2 = divide_numbers(10, 0)
print(f"Result of 10 / 0: {result2}")
```

### Question 2(a): List and Dictionary Comprehensions

```python

# Solution for Question 2(a)
words = ["data", "science", "python", "numpy", "pandas", "visualization"]

# List of word lengths
word_lengths = [len(word) for word in words]
print("Word lengths:", word_lengths)

# Dictionary of words and their lengths
word_dict = {word: len(word) for word in words}
print("Word dictionary:", word_dict)

# Filtered list for words with more than 5 letters
filtered_words = [word for word in words if len(word) > 5]
print("Filtered words:", filtered_words)
```

### Question 2(b): Tuple and Set Operations

```python

# Solution for Question 2(b)
platform_a_users = ("alice", "bob", "charlie", "diana")
platform_b_users = ("charlie", "diana", "edward", "fiona")

# Convert tuples to sets
set_a = set(platform_a_users)
set_b = set(platform_b_users)

# Users registered on both platforms
common_users = set_a.intersection(set_b)
print("Common users:", common_users)

# Users only on Platform A
unique_to_a = set_a.difference(set_b)
print("Users only on Platform A:", unique_to_a)

# Combined unique users
all_users = tuple(set_a.union(set_b))
print("All unique users:", all_users)
```

### Question 4(a): Data Loading and Basic Exploration

```python

# Solution for Question 4(a)
import pandas as pd

# Load dataset
movies_df = pd.read_csv('movies.csv')

# Display first 10 rows
print("First 10 rows:\n", movies_df.head(10))

# Summary of the DataFrame
print("Summary of DataFrame:\n", movies_df.info())

# Handle missing values in 'Rating'
movies_df['Rating'] = movies_df['Rating'].fillna(movies_df['Rating'].mean())
print("Missing values in 'Rating' handled.")
```

### Question 4(b): Data Filtering and Grouping

```python

# Solution for Question 4(b)
# Filter movies released after 2000 with rating > 8.0
filtered_movies = movies_df[(movies_df['ReleaseYear'] > 2000) & (movies_df['Rating'] > 8.0)]
print("Filtered movies:\n", filtered_movies)

# Group by Genre and calculate average Rating
genre_avg_rating = movies_df.groupby('Genre')['Rating'].mean()
print("Average rating by genre:\n", genre_avg_rating)

# Top 3 directors with highest average ratings
top_directors = movies_df.groupby('Director')['Rating'].mean().nlargest(3)
print("Top 3 directors by average rating:\n", top_directors)
```

### Question 4(c): Data Visualization

```python

# Solution for Question 4(c)
import matplotlib.pyplot as plt

# Movies released each year from 2000 to 2023
movies_per_year = movies_df[movies_df['ReleaseYear'] >= 2000].groupby('ReleaseYear').size()

# Bar chart
movies_per_year.plot(kind='bar', title='Movies Released Per Year (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```

### Question 5(a): Matplotlib Visualization

```python

# Solution for Question 5(a)
# Scatter plot for ReleaseYear vs Rating
plt.figure(figsize=(10, 6))
for genre in filtered_movies['Genre'].unique():
    genre_data = filtered_movies[filtered_movies['Genre'] == genre]
    plt.scatter(genre_data['ReleaseYear'], genre_data['Rating'], label=genre)

plt.title('Release Year vs Rating by Genre')
plt.xlabel('Release Year')
plt.ylabel('Rating')
plt.legend(title='Genre')
plt.show()
```

### Question 5(b): Seaborn Visualization

```python

# Solution for Question 5(b)
import seaborn as sns

# Facet grid of histograms for movie ratings by genre
sns.displot(data=movies_df, x='Rating', col='Genre', kde=True, col_wrap=3, height=3)
plt.suptitle('Distribution of Movie Ratings by Genre', y=1.02)
plt.show()
```

### Question 6: Hands-On Project

```python

# Solution for Question 6
# Load ecommerce data
ecommerce_df = pd.read_csv('ecommerce_data.csv')
ecommerce_df['OrderDate'] = pd.to_datetime(ecommerce_df['OrderDate'])

# Add TotalRevenue column
ecommerce_df['TotalRevenue'] = ecommerce_df['Quantity'] * ecommerce_df['UnitPrice']

# Top 5 customers by total revenue
top_customers = ecommerce_df.groupby('CustomerID')['TotalRevenue'].sum().nlargest(5)
print("Top 5 customers by total revenue:\n", top_customers)

# Pie chart for top 5 customers
top_customers.plot(kind='pie', title='Top 5 Customers by Revenue', autopct='%1.1f%%')
plt.ylabel('')
plt.show()

# Line plot for monthly total revenue
monthly_revenue = ecommerce_df.resample('M', on='OrderDate')['TotalRevenue'].sum()
monthly_revenue.plot(title='Monthly Total Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

### Question 7: Capstone Coding Challenge

```python

# Solution for Question 7
# Load stock price data
stock_data = pd.read_csv('stock_prices.csv')

# Convert 'Date' to datetime and set as index
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.set_index('Date', inplace=True)

# Calculate 7-day moving average
stock_data['7-Day MA'] = stock_data['ClosePrice'].rolling(window=7).mean()

# Detect and replace outliers
from scipy.stats import zscore
z_scores = zscore(stock_data['ClosePrice'].fillna(method='ffill'))
outliers = abs(z_scores) > 3
stock_data.loc[outliers, 'ClosePrice'] = stock_data['ClosePrice'].median()

# Plot original and moving average prices
plt.figure(figsize=(10, 6))
plt.plot(stock_data.index, stock_data['ClosePrice'], label='Original Prices', alpha=0.7)
plt.plot(stock_data.index, stock_data['7-Day MA'], label='7-Day MA', color='red')
plt.title('Stock Prices and Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
