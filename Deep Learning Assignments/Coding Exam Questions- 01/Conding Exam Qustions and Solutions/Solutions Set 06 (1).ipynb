### Question 1(b): Control Flow and Functions

```python

# Solution for Question 1(b)
def fibonacci_sequence(n):
    if not isinstance(n, int) or n <= 0:
        raise ValueError("Input must be a positive integer.")
    sequence = [0, 1]
    while len(sequence) < n:
        sequence.append(sequence[-1] + sequence[-2])
    return sequence[:n]

try:
    print("Fibonacci sequence (n=10):", fibonacci_sequence(10))
except ValueError as e:
    print(e)
```

### Question 2(a): List and Dictionary Manipulations

```python

# Solution for Question 2(a)
students = [
    {"name": "Alice", "score": 88},
    {"name": "Bob", "score": 75},
    {"name": "Charlie", "score": 92},
    {"name": "Diana", "score": 85},
    {"name": "Evan", "score": 79}
]

# Sort by scores in ascending order
sorted_students = sorted(students, key=lambda x: x['score'])
print("Sorted students by score:", sorted_students)

# Create dictionary mapping names to scores
student_dict = {student['name']: student['score'] for student in students}
print("Dictionary of students and their scores:", student_dict)
```

### Question 2(b): Set Operations

```python

# Solution for Question 2(b)
developer_1 = {"Python", "Java", "C++", "JavaScript"}
developer_2 = {"Ruby", "JavaScript", "Python", "Go"}

# Languages known by both developers
common_languages = developer_1.intersection(developer_2)
print("Languages known by both developers:", common_languages)

# Languages known by developer_1 but not developer_2
unique_languages_1 = developer_1.difference(developer_2)
print("Languages known only by developer_1:", unique_languages_1)

# All languages known by either developer
all_languages = developer_1.union(developer_2)
print("All languages known by either developer:", all_languages)
```

### Question 3(a): Array Creation and Mathematical Operations

```python

# Solution for Question 3(a)
import numpy as np

# Create array A
A = np.arange(1, 10).reshape(3, 3)
print("Matrix A:\n", A)

# Compute the inverse
try:
    A_inv = np.linalg.inv(A)
    print("Inverse of A:\n", A_inv)

    # Verify the dot product gives the identity matrix
    identity_matrix = np.dot(A, A_inv)
    print("Dot product of A and its inverse:\n", identity_matrix)
except np.linalg.LinAlgError as e:
    print(f"Error: {e}")
```

### Question 3(b): Statistical Functions

```python

# Solution for Question 3(b)
scores = np.array([88, 75, 92, 85, 79, 95, 68, 74, 81, 90])

# Statistics
mean_score = np.mean(scores)
median_score = np.median(scores)
std_dev_score = np.std(scores)
percentage_above_mean = (np.sum(scores > mean_score) / scores.size) * 100

print(f"Mean: {mean_score}, Median: {median_score}, Std Dev: {std_dev_score}")
print(f"Percentage of students scoring above the mean: {percentage_above_mean:.2f}%")
```

### Question 4(a): Data Loading and Cleaning

```python

# Solution for Question 4(a)
import pandas as pd

# Load the dataset
transactions_df = pd.read_csv('transactions.csv')

# Convert 'TransactionDate' to datetime format and extract 'Month'
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
transactions_df['Month'] = transactions_df['TransactionDate'].dt.month

# Check for missing values and fill with appropriate values
transactions_df.fillna(transactions_df.mean(numeric_only=True), inplace=True)
print("Handled missing values.")
```

### Question 4(b): Data Aggregation and Grouping

```python

# Solution for Question 4(b)
# Calculate total revenue for each product
transactions_df['TotalRevenue'] = transactions_df['Quantity'] * transactions_df['UnitPrice']
product_revenue = transactions_df.groupby('Product')['TotalRevenue'].sum()

# Identify top 5 products by total revenue
top_5_products = product_revenue.nlargest(5)
print("Top 5 products by total revenue:\n", top_5_products)

# Total quantity sold for each category
category_quantity = transactions_df.groupby('Category')['Quantity'].sum()
print("Total quantity sold by category:\n", category_quantity)
```

### Question 4(c): Data Visualization

```python

# Solution for Question 4(c)
import matplotlib.pyplot as plt

# Bar chart for top 5 products by revenue
top_5_products.plot(kind='bar', color='skyblue', title='Top 5 Products by Revenue')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.show()
```

### Question 5(a): Time Series Visualization

```python

# Solution for Question 5(a)
# Calculate monthly total revenue
monthly_revenue = transactions_df.resample('M', on='TransactionDate')['TotalRevenue'].sum()

# Plot line chart for monthly revenue
plt.figure(figsize=(10, 6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o')
plt.title('Monthly Total Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

### Question 5(b): Advanced Visualization

```python

# Solution for Question 5(b)
# Box plot for unit prices by category
import seaborn as sns

sns.boxplot(x='Category', y='UnitPrice', data=transactions_df)
plt.title('Unit Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Unit Price')
plt.show()
```

### Question 6: Hands-On Project

```python

# Solution for Question 6
# Load employee attendance data
attendance_df = pd.read_csv('employee_attendance.csv')
attendance_df['Date'] = pd.to_datetime(attendance_df['Date'])

# Calculate attendance percentage
attendance_percentage = attendance_df.groupby('EmployeeID')['Status'].apply(lambda x: (x == 'Present').mean() * 100)
print("Attendance percentage:\n", attendance_percentage)

# Department with highest average attendance
department_attendance = attendance_df.groupby('Department')['Status'].apply(lambda x: (x == 'Present').mean() * 100)
top_department = department_attendance.idxmax()
print(f"Department with highest average attendance: {top_department}")

# Bar chart for top 5 employees by attendance
top_5_employees = attendance_percentage.nlargest(5)
top_5_employees.plot(kind='bar', title='Top 5 Employees by Attendance')
plt.xlabel('Employee ID')
plt.ylabel('Attendance Percentage')
plt.show()
```

### Question 7: Capstone Coding Challenge

```python

# Solution for Question 7
from collections import Counter
import string

# Read and process reviews
with open('reviews.txt', 'r') as file:
    reviews = file.readlines()

# Clean reviews
cleaned_reviews = [review.translate(str.maketrans('', '', string.punctuation)).lower() for review in reviews]

# Tokenize and count positive/negative words
positive_words = ['good', 'great', 'excellent', 'amazing', 'fantastic', 'love']
negative_words = ['bad', 'poor', 'terrible', 'awful', 'hate', 'worst']

positive_count = sum(word in positive_words for review in cleaned_reviews for word in review.split())
negative_count = sum(word in negative_words for review in cleaned_reviews for word in review.split())

# Sentiment score
sentiment_score = positive_count - negative_count
print(f"Sentiment Score: {sentiment_score}")

# Pie chart for word proportions
plt.pie([positive_count, negative_count], labels=['Positive Words', 'Negative Words'], autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Proportion of Positive and Negative Words')
plt.show()
```
