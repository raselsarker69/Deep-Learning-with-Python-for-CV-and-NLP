3. Image Conversion

```python
import cv2
import matplotlib.pyplot as plt

# Corrected image path with file extension
image_path = "/kaggle/input/animal/Dataset/test/cat/6fcc622575e0d1fd0cf7f04c4472fd6a0f97defb7cfc3bc54e2e9a7caa9ddc438c190d7_1920.png"

# Read the image
image = cv2.imread(image_path)

# Check if image was loaded successfully
if image is None:
    print("Error: Image not found or path is incorrect.")
else:
    # Convert BGR to RGB for proper visualization
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Convert to grayscale using OpenCV
    gray_image_opencv = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Convert to grayscale manually using weighted sum (fixed line break)
    gray_manual = (
        0.2989 * rgb_image[:, :, 0] +  # Red channel
        0.5870 * rgb_image[:, :, 1] +  # Green channel
        0.1140 * rgb_image[:, :, 2]    # Blue channel
    )
    
    # Plot the images
    plt.figure(figsize=(12, 6))
    
    # OpenCV Grayscale
    plt.subplot(1, 2, 1)
    plt.imshow(gray_image_opencv, cmap="gray")
    plt.title("Grayscale (OpenCV)")
    plt.axis("off")
    
    # Manual Grayscale
    plt.subplot(1, 2, 2)
    plt.imshow(gray_manual, cmap="gray")
    plt.title("Grayscale (Manual)")
    plt.axis("off")
    
    plt.show()
```
